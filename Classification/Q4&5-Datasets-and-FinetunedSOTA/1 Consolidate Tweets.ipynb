```python
#import libraries
import pandas as pd
```

```python
df = pd.concat(map(pd.read_csv, ['Crawled Tweets/covid_vaccine.csv', 'Crawled Tweets/moderna_vaccine.csv', 'Crawled Tweets/pfizer_vaccine.csv']), ignore_index=True)
print("shape:",df.shape)
df.tail()
```

```python
df2 = pd.read_csv('Crawled Tweets/sinovac_vaccine.csv')
print("shape:",df2.shape)
df2.tail()
```

```python
df = pd.concat([df, df2[['created_at', 'id_str', 'full_text', 'quote_count', 'reply_count',
       'retweet_count', 'favorite_count', 'lang', 'user_id_str',
       'conversation_id_str', 'username', 'tweet_url']]])
print("shape:",df.shape)
df.tail()
```

```python
old = df.shape[0]
df.drop_duplicates(subset=['created_at', 'id_str', 'full_text', 'quote_count', 'reply_count',
       'retweet_count', 'favorite_count', 'lang', 'user_id_str',
       'conversation_id_str', 'username', 'tweet_url'], inplace=True)
print("Number of duplicate tweets dropped:",old - df.shape[0])
```

```python
df.to_csv('tweet_corpus.csv', index=False)  
```
